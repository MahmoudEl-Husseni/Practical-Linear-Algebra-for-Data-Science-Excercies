# Exercise 3-1
> Rewrite the code for linear weighted combination, but put the scalars in a list and the vectors as elements in a list (thus, you will have two lists, one of scalars and one of NumPy arrays). Then use a for loop to implement the linear weighted combination operation. Initialize the output vector using np.zeros(). Confirm that you get the same result as in the previous code.

```python
# Using Separate Variables 
l1 = 1
l2 = 2
l3 = -3
v1 = np.array([4,5,1])
v2 = np.array([-4,0,-4])
v3 = np.array([1,3,2])
l = l1*v1 + l2*v2 + l3*v3

# Using list
weights = [1, 2, -3]
vectors = [v1, v2, v3]
weighted_vector = np.zeros_like(v1)

for w, v in zip(weights, vectors): 
    weighted_vector += w*v

print(f"Weighted vector using Separate variables method: ".ljust(50), f"{red}{l}{res}")
print(f"Weighted vector using Loop: ".ljust(50), f"{red}{weighted_vector}{res}")
```

# Exercise 3-2.
> Although the method of looping through lists in the previous exercise is not as efficient as matrix-vector multiplication, it is more scalable than without a for loop. You can explore this by adding additional scalars and vectors as elements in the lists. What happens if the new added vector is in ℝ4 instead of ℝ3? And what happens if you have more scalars than vectors?

```python
# Adding Additional scalars
weights = np.arange(10)
vectors = np.random.randint(0, 10, (10, 3))

weighted_vector = np.zeros_like(vectors[0])
for w, v in zip(weights, vectors): 
    weighted_vector += w*v

print(f"Adding additional scalars: ".ljust(50), f"{red}{weighted_vector}{res}")
print(f"{blk}{70*'='}{res}")


# Vectors in R4
vectors = np.random.randint(0, 5, (10, 4))
weighted_vector = np.zeros_like(vectors[0])
for w, v in zip(weights, vectors): 
    weighted_vector += w*v

print(f"Using Vectors in R4: ".ljust(50), f"{red}{weighted_vector}{res}")
print(f"{blk}{70*'='}{res}")


# More Scalars than vectors: 
weights = np.arange(20)
weighted_vector = np.zeros_like(vectors[0])
for w, v in zip(weights, vectors): 
    weighted_vector += w*v

print(f"More scalars than vectors: ".ljust(50), f"{red}{weighted_vector}{res}")
print(f"{blk}{70*'='}{res}")
```

- Adding additional scalars & vectors: doesn't matter as they are stored in list
- Using Vectors in R4: output vector will be in R4
- Using more scalars than vectors: loop will just iterates over number of vectors and ignores rest of scalars

# Exercise 3-3.
> In this exercise, you will draw random points in subspaces. This will help reinforce the idea that subspaces comprise any linear weighted combination of the spanning vectors. Define a vector set containing one vector [1, 3]. Then create 100 numbers drawn randomly from a uniform distribution between −4 and +4. Those are your random scalars. Multiply the random scalars by the basis vector to create 100 random points in the subspace. Plot those points.

```python
basis = np.array([[1, 3]])
rand_weights = np.random.uniform(-4, 4, (1, 100))


points = basis.T.dot(rand_weights)
points = points.T



fig = go.Figure()

fig.add_trace(go.Scatter(
    x=points[:,0], y=points[:,1], mode='markers' ))

fig.show()
```

```python
basis = np.array([[3, 5, 1], [0, 2, 2]]).reshape(2, -1)
rand_weights = np.random.uniform(-4, 4, (100, 2)).reshape(2, -1)


points = basis.T.dot(rand_weights)
points = points.T



fig = go.Figure()

fig.add_trace(go.Scatter3d(
    x=points[:,0], y=points[:,1], z=points[:,2], mode='markers' ))

fig.show()
```
